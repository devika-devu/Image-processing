### Import necessary Libraries

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

### Read 'image1.jpg' with opencv

```python
img = cv2.imread('image1.jpg')
```

```python
print(img)
```

### Plot the image with matplotlib

```python
plt.imshow(img)
```

### Convert the image to RGB

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
resize_img=cv2.resize(img_rgb,(300,400))
plt.imshow(resize_img)
```

### Check the shape of the image

```python
img_rgb.shape
```

### Resize the image to (400,300)

```python
img=cv2.resize(img_rgb,(400,300))
plt.imshow(img)
```

### Resize the image by ratio. Use 0.5 height and width

```python
w_ratio = 0.5
h_ratio = 0.5
```

```python
new_img = cv2.resize(img_rgb, (0,0), img, w_ratio, h_ratio)
```

```python
plt.imshow(new_img)
```

### Flip the image along central x axis

```python
new_img = cv2.flip(new_img, 0)
plt.imshow(new_img)
```

## Flip the image along central y axis


```python
new_img = cv2.flip(new_img, 1)
plt.imshow(new_img)
```

### Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### Open 'image2.jpg' from the data folder using open cv and display it in the notebook. Make sure to correct for the RGB order

```python
img=cv2.imread('image2.jpg')
```

```python
img
```

##### Before RGB correction

```python
plt.imshow(img)
```

### After correction

```python
img_rgh=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
img_rgh
```

```python
plt.imshow(img_rgh)
```

### Draw an empty RED rectangle around toms face and display the image in the notebook

```python
draw_rect=cv2.rectangle(img_rgh, pt1 = (300,400), pt2 = (600,700), color = (255,0,0), thickness = 5)
```

```python
plt.imshow(draw_rect)
```

### Draw a text named 'TOM' above toms face

```python
font=cv2.FONT_HERSHEY_SIMPLEX
```

```python
draw_text=cv2.putText(draw_rect,text='TOM',org=(500,770),fontFace=font,fontScale=3,color=(0,0,0),thickness=12,lineType=cv2.LINE_AA )
```

```python
plt.imshow(draw_text)
```

### Do the same steps for 'Jerry'

```python
draw_rect2=cv2.rectangle(img_rgh, pt1 = (100,500), pt2 = (270,700), color = (255,0,0), thickness = 5)
```

```python
plt.imshow(draw_rect2)
```

```python
font=cv2.FONT_HERSHEY_SIMPLEX
```

```python
draw_text1=cv2.putText(draw_rect,text='JERRY',org=(120,750),fontFace=font,fontScale=2,color=(0,0,0),thickness=10,lineType=cv2.LINE_AA )
```

```python
plt.imshow(draw_text1)
```

### Convert the image to grayscale

```python
figure =  cv2.cvtColor(draw_text1, cv2.COLOR_RGB2GRAY)
```

```python
plt.imshow(figure,cmap='gray')
```

### Display the image in a larger size

```python
plt.figure(figsize = (10,10))
figure = cv2.cvtColor(draw_text1, cv2.COLOR_RGB2GRAY)
plt.imshow(figure,cmap='gray')
```

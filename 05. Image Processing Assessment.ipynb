### Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### Create a function to display the images in larger size

```python
def display_img(img, cmap = None):
    plt.figure(figsize = (8,8))
    plt.imshow(img, cmap)
```

### Read 'panda.jpg' using OpenCV and convert to RGB`M

```python
img=cv2.imread('panda.jpg')
```

```python
plt.imshow(img)
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
display_img(img)
```

### Apply a binary threshold onto the image. 

```python
ret,thresh = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
```

```python
ret
```

```python
new=cv2.cvtColor(thresh,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(new,cmap='gray')
```

### Convert colorspace to HSV and display the image

use 'cv2.COLOR_BGR2HSV'

```python
img_new=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
```

```python
plt.imshow(img_new)
```

### Create a low pass filter with a 4 by 4 Kernel filled with values of 1/10 (0.01) and then use 2-D Convolution to blur the giraffer image (displayed in normal RGB)

```python
kernel= np.ones((4,4))*0.1
kernel
```

```python
dst=cv2.filter2D(img,-1,kernel)
display_img(dst)
```

### Create a Horizontal Sobel Filter (sobelx) with a kernel size of 5 to the grayscale version of the giaraffes image and then display the resulting gradient filtered version of the image.

```python
img_gray=cv2.imread('panda.jpg',0)
```

```python
sobelx= cv2.Sobel(img_gray,cv2.CV_64F,1,0,ksize=5)
```

```python
display_img(sobelx, cmap = 'gray')
```

Plot the color histograms for the RED, BLUE, and GREEN channel of the giaraffe image. Pay careful attention to the ordering of the channels.

```python
img =cv2.imread('panda.jpg')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.title('Panda Hist')
plt.show()
```

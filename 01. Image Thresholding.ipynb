Thresholding is a type of image segmentation, where we change the pixels of an image to make the image easier to analyze. In thresholding, we convert an image from color or grayscale into a binary image, i.e., one that is simply black and white. Most frequently, we use thresholding as a way to select areas of interest of an image, while ignoring the parts we are not concerned with.

### Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

### Read 'image4.jpg' with cv2

```python
img = cv2.imread('image4.jpg')
```

```python
plt.imshow(img)
```

### Adding the 0 flag to read it in black and white

```python
img = cv2.imread('image4.jpg', 0)
```

```python
plt.imshow(img, cmap = 'gray')
```

## Different Threshold Types

### Binary

Creates a binary image from a gray image based on the threshold values. The object target can be specified as dark or light.

Any values below 127 will be set to 0 & any value after 127 will be set to max

```python
ret, thresh1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
```

```python
ret
```

```python
plt.imshow(thresh1, cmap = 'gray')
```

### Binary Inverse

```python
ret, thresh2 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY_INV)
```

```python
plt.imshow(thresh2, cmap ='gray')
```

### Threshold Truncation

```python
ret, thresh3 = cv2.threshold(img, 127, 255, cv2.THRESH_TRUNC)
plt.imshow(thresh3, cmap = 'gray')
```

### Threshold to Zero

```python
ret, thresh4 = cv2.threshold(img, 127, 255, cv2.THRESH_TOZERO)
plt.imshow(thresh4, cmap = 'gray')
```

### Threshold to Zero (Inverse)

```python
ret, thresh5 = cv2.threshold(img, 127,255, cv2.THRESH_TOZERO_INV)
plt.imshow(thresh5, cmap= 'gray')
```

### Import necessary libraries

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Read 'chess.jpg'

```python
img = cv2.imread('chess.jpg')
```

### View the image with matplotlib

```python
plt.imshow(img)
```

### Increase brightness by gamma correction

```python
def load_img():
    img = cv2.imread('chess.jpg').astype(np.float32) / 255
    return img
```

```python
def display_img(img):
    fig = plt.figure(figsize=(12,10))
    ax = fig.add_subplot(111)
    ax.imshow(img)
```

```python
img = load_img()
gamma = 1/8
effected_image = np.power(img, gamma)
display_img(effected_image)
```

### Apply Gaussian Blur on the original image

```python
img = load_img()
blurred_img=cv2.GaussianBlur(img,(5,5),10)
plt.imshow(blurred_img)
```

### Apply median Blurring

```python
median=cv2.medianBlur(img,5)
plt.imshow(median)
```

### Apply bilateral Filtering

```python
img = load_img()
display_img(img)
```
